```python
%matplotlib notebook
from matplotlib.widgets import Button
import matplotlib.pyplot as plt
def callback(event):
    plt.text(event.xdata, event.ydata, 'clicked\n'+ 'x:'+str(event.xdata) + '\ny:' +str(event.ydata))

f,a = plt.subplots(1)
b1 = Button(a,'Button1')
b1.on_clicked(callback)
plt.show()
```
