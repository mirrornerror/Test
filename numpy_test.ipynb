```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
plt.style.use("ggplot")
```

```python
mu = 0
sigma = 1.0
num = 100
n = 10
#x = np.random.normal(mu,sigma,num)
#y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2))
x = norm.rvs(mu, sigma, num)
y = norm.pdf(x, mu, sigma)

xx = np.array(np.zeros([num,n]))
yy = np.array(np.zeros([num,n]))

for i in range(num):
    xx[i] = x[i]
    yy[i] = norm.rvs(0, 0.03, n) + y[i]

plt.xlim([-4,4])
plt.scatter(xx, yy, c="#666666", s=2, alpha=0.3)
plt.scatter(x, y, c="#ff0000", s=4)
```
