```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
plt.style.use("ggplot")
def plot(X,Y):
    plt.axes().set_aspect('equal')
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.plot(X,Y)
x = np.linspace(-8,8,100)
```

## $y = ax + b$ 

```python
y = 0.5*x + 2
plot(x,y)
```

```python
y = x**2
plot(x,y)
```

```python
y = x**2*0.05
plot(x,y)
```

## $y = ax^2 + bx + c$

```python
y = 0.05*x**2 + 0.2*x + 1
plot(x,y)
```

```python
y = 0.05*x**3
plot(x,y)
```

```python
y = -x*np.sin(x)
the = 0.1
xx = x*np.cos(the) - y*np.sin(the)
yy = x*np.sin(the) + y*np.cos(the)
plot(xx,yy)
```

```python
# sigmoid
a = 80
y = 1/(1+np.exp(-a*x))*5
plot(x,y)
```

```python
r = 6
y = np.ma.sqrt(r**2 - x**2)
plot(x,y)
```

```python
r = 6
the = np.linspace(-np.pi,np.pi,100)
x = r*np.cos(the)
y = r*np.sin(the)
plot(x,y)
```

```python
y = np.ma.log(1+np.ma.exp(x)) # softplus
plot(x,y)
```

```python
y = (x>0)*x # relu
plot(x,y)
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)

def sine(a,n): 
    for i in range(n):
        a = np.sin(a)
    return a

ys = sine(x,10000)*50

plt.xlim(0,10)
# plt.ylim(-3,3)



plt.plot(x,y,'r--',alpha=0.3)
plt.plot(x,ys,'b-')
```

```python
from numpy.fft import fft 

N = 200
# sample spacing
T = 1.0 / 400.0
x = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x) + np.sin(x*100)
plt.plot(x,y,'r-')
```

```python
#y = np.sin(x*100)
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
plt.plot(xf, np.abs(yf[0:N//2]))
np.argmax(np.abs(yf[0:N//2]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import norm,beta,multivariate_normal
# %matplotlib ipympl
%matplotlib inline
# %matplotlib nbagg

x = np.arange(-2,2,0.1)
y = np.arange(-2,2,0.1)

X, Y = np.meshgrid(x, y)
Z = norm.pdf(X,loc=1.8,scale=0.2) * norm.pdf(Y,loc=1.8,scale=0.2) \
    + norm.pdf(X,loc=1.8, scale=0.2) * norm.pdf(Y,loc=-1.8,scale=0.2) \
    + norm.pdf(X,loc=-1.8,scale=0.2) * norm.pdf(Y,loc=1.8,scale=0.2) \
    + norm.pdf(X,loc=-1.8,scale=0.2) * norm.pdf(Y,loc=-1.8,scale=0.2) \
    - norm.pdf(Y,loc=1.8, scale=0.2) * beta.pdf(x, a=1, b=1, loc=-1.25,scale=2.5)*5

fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(111, projection='3d')
ax.set_aspect(1.8)
#ax.scatter3D(X,Y,-Z,'r.')
ax.plot_wireframe(X,Y,-Z)
#ax.plot_surface(X,Y,-Z,color='r')
plt.show()
```
