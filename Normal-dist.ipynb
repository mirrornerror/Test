```python
from scipy.stats import norm
```

```python
num = 100

mu1 = 0.0
sig1 = 1.0
x1 = norm.rvs(loc=mu1, scale=sig1, size=num)
y1 = norm.pdf(x1, loc=mu1, scale=sig1)

mu2 = 2
sig2 = 1.5
x2 = norm.rvs(loc=mu2, scale=sig2, size=num )
y2 = norm.pdf(x2, loc=mu2, scale=sig2)

mu3 = (mu1 + mu2)/2
sig3 = (sig1**2 + sig2**2)**0.5
#sig3 = (sig1+sig2)
x3 = np.concatenate([x1,x2])
y3 = norm.pdf(x3, loc=mu3, scale=sig3)
```

```python
plt.plot(x1,y1,'.', label='x1')
plt.plot(x2,y2,'.', label='x2')
plt.plot(x3,y3,'.', label='x3')
plt.legend()
print('mu3:  ',mu3)
print('sig3: ',sig3)
```
